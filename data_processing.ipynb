```python
GAME_WINDOW = 20 # measure 20 games into te past
TEAM_SIZE = 10 # take in 10 players per team per input

player_stats = {}
for index, row in df.iterrows():
    player_name = row["PLAYER_NAME"]
    
    stats = [row["GAME_DATE"]] + row.iloc[7:28].to_list()
    stats[1] = 1 if stats[1] == "W" else 0

    if player_name not in player_stats.keys():
        player_stats[player_name] = [stats]
    else:
        # sorted insert!!!!!
        #position = bisect.bisect_left(player_stats, )
        player_stats[player_name].append(stats)


for key, games in player_stats.items():
    print(key, games)
    #print(len(value))
```

```python
#PLAYER_NAME : GAME_DATE(0),WL(1),MIN(2),FGM(3),FGA(4),(calculate this with FGM/FGA)(5),FG3M,FG3A,(calculate this with FG3M/FG3A)(8),FTM,FTA,(calculate this with FTM/FTA)(11),OREB,DREB,REB,AST,STL,BLK,TOV,PF,PTS,PLUS_MINUS
player_pre_stats = {}
for name, games in player_stats.items():
    i = 0
    for game in games:
        if i == 0:
            player_pre_stats[name] = [[game[0]] + [0] * 21] 
        elif i <= 20: # consider lagging behind 1 so that i wouldn't have to do games[i-1][j] so often
            pre_stats = [game[0]] + [0] * 21
            dummy_stats = player_pre_stats[name][i-1]

            # for W/L
            pre_stats[1] = dummy_stats[1] + games[i-1][1]

            for j in range(2, 22):
                # so that season field goal percentages aren't skewed by low volume performances
                if j == 5 or j == 8 or j == 11: 
                    made = pre_stats[j-2]
                    attempted = pre_stats[j-1]
                    pre_stats[j] = (0 if attempted == 0 else made / attempted)
                else:
                    pre_stats[j] = (dummy_stats[j] * (i - 1) + games[i-1][j]) / (i)

            player_pre_stats[name].append(pre_stats)
        else:
            pre_stats = [game[0]] + [0] * 21
            dummy_stats = player_pre_stats[name][i-1]

            # for W/L
            pre_stats[1] = dummy_stats[1] + games[i-1][1] - games[i-GAME_WINDOW][1]

            for j in range(2, 22):
                if j == 5 or j == 8 or j == 11:
                    made = pre_stats[j-2]
                    attempted = pre_stats[j-1]
                    pre_stats[j] = (0 if attempted == 0 else made / attempted)
                else:
                    pre_stats[j] = (dummy_stats[j] * (i - 1) + games[i-1][j] - games[i-GAME_WINDOW][j]) / (i)
            
            player_pre_stats[name].append(pre_stats)

        i += 1

for name, until_game_stats_list in player_pre_stats.items():
    print(name, until_game_stats_list)
```

```python
# dictionary of a dictionary of a list
# player_name -> game_date -> game_stats
player_stats_dict = {}
for name, until_game_stats_list in player_pre_stats.items():
    player_stats_dict[name] = {}
    # focusing on a single player here
    for until_game_stats in until_game_stats_list:
        game_date = until_game_stats[0]
        player_stats_dict[name][game_date] = until_game_stats[1:]

for name, dict_stats in player_stats_dict.items():
    print(name)
    for date, stats in dict_stats.items():
        print(date, stats)
```
